```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

import numpy as np
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris.data[:5]
```

```python
iris.target[:5]
```

```python
X = iris.data
y = iris.target
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# knn_clf = KNeighborsClassifier()
# knn_clf.fit(X, y)
```

## 1. K- 폴드 교차 검증

```python
from sklearn.metrics import accuracy_score # 정확도
```

```python
accuracy_list = []

knn_clf = KNeighborsClassifier()

kfold = KFold(n_splits=5, shuffle=True)
for train_index, valid_index in kfold.split(X):
  # 훈련(학습)
  X_train, y_train = X[train_index], y[train_index] # 4/5
  X_valid, y_valid = X[valid_index], y[valid_index] # 1/5
  knn_clf.fit(X_train, y_train)
  
  
  # 예측과 평가(정확도)
  pred = knn_clf.predict(X_valid)
  accuracy = accuracy_score(y_valid, pred)
  accuracy_list.append(accuracy)

print("평균 정확도 : ", np.mean(accuracy_list))
```

## 2. Stratified K 폴드 교차 검증

```python
accuracy_list = []

knn_clf = KNeighborsClassifier()

skfold = StratifiedKFold(n_splits=5, shuffle=True)
for train_index, valid_index in skfold.split(X, y):
  # 훈련(학습)
  X_train, y_train = X[train_index], y[train_index] # 4/5
  X_valid, y_valid = X[valid_index], y[valid_index] # 1/5
  knn_clf.fit(X_train, y_train)
  
  
  # 예측과 평가(정확도)
  pred = knn_clf.predict(X_valid)
  accuracy = accuracy_score(y_valid, pred)
  accuracy_list.append(accuracy)

print("평균 정확도 : ", np.mean(accuracy_list))
```

## 3. cross_val_score()

```python
from sklearn.model_selection import cross_val_score

knn_clf = KNeighborsClassifier()

#kfold = KFold(n_splits=5, shuffle=True)
skfold = StratifiedKFold(n_splits=5, shuffle=True)

scores = cross_val_score(knn_clf, X, y, scoring="accuracy", cv=skfold) # scoring은 검증 데이터를 어떤 성능 지표로 평가할지
```
